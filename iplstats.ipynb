```python
import pandas as pd
```

```python
updated_balls = pd.read_csv('2008_2021_updated_ball.csv', usecols = ['id','batsman','batsman_runs'])
updated_balls.head()
```

```python
data_matches = pd.read_csv('2008_2021_data_matches.csv', usecols = ['id','season'])
data_matches.head()
```

```python
data_merged = updated_balls.merge(data_matches, on='id', how='left')
data_merged.head()
```

# Overall Stats

```python
data_merged.query('season != "2022"').groupby(
    ['batsman'], as_index=False).agg(
        {"id":"nunique", "batsman_runs":"sum"}).sort_values(
            ['batsman_runs'],ascending=False).reset_index(
                drop=True)
```

# Seasonal Stats

```python
data_merged.query('season != "2022"').groupby(
    ['season','batsman'], as_index=False).agg(
        {"id":"nunique", "batsman_runs":"sum"}).sort_values(
            ['batsman_runs'],ascending=False).drop_duplicates(
                subset = 'season').sort_values(
                    ['season']).reset_index(
                        drop=True)
```
